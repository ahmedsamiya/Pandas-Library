```python
import pandas as pd
```

```python
#merge operation perform based on common column
#join operation perform based on index
```

```python
#join
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
                
```

```python
df2 =  pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
df1,df2
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='cross')
```

```python
df1,df2
```

```python
pd.merge(df1,df2,how='right',left_on='key3',right_on='key5')
```

```python
def create_flag(x):
    if x<10:
        return "cheap"
    elif x>10 and x<90:
        return "medium"
    else:
        return "high"
    
```
